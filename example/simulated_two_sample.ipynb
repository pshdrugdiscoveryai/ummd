```python
import time
import scipy.spatial.distance as distance
import numpy as np
```

```python
from ummd import mmd
from ummd.kernels import GaussianKernel
```

```python
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()
        wrapper.time_taken = end - start
        return res
    return wrapper
```

```python
x = np.random.normal(size=(10, 1), loc=0, scale=1)
y = np.random.normal(size=(9, 1), loc=1, scale=0.5)
```

```python
@timer
def time_function(func, *args):
    result = func(*args)

time_function(mmd.MMD, x, y)
print(f"Time taken: {time_function.time_taken:.4f} seconds")
```
